```python
#combine all the filters together
```

```python
df2 = df[(df['Event distance/length'].isin(['50km', '50mi'])) & 
         (df['Year of event'] == 2020) & 
         (df['Event name'].str.split('(').str.get(1).str.split(')').str.get(0) == 'USA')]
```

```python
df2.shape
```

```python
#remove (USA) from event name
```

```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

```python
df2.head()
```

```python
#clean up the athlete age
```

```python
df2['athlete_age'] = 2020 - df2['Athlete year of birth']
```

```python
#remove h from athlete performance
```

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
```

```python
df2.head(5)
```

```python
#drop columns: Athlete Club, Athlete Country, Athlete year of birth, Athlete Age Category
```

```python
df2 = df2.drop(columns=['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'])
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
#rename columns
```

```python
df2 = df2.rename(columns={
    'Year of event': 'year',
    'Event dates': 'race_day',
    'Event name': 'race_name',
    'Event distance/length': 'race_length',
    'Event number of finishers': 'race_number_of_finishers',
    'Athlete performance': 'athlete_performance',
    'Athlete gender': 'athlete_gender',
    'Athlete average speed': 'athlete_average_speed',
    'Athlete ID': 'athlete_id'
})
```

```python
#questions I want to find out from the data
```

```python
#Difference in speed for the 50k,50mi male to female
```

```python
df2.groupby(['race_length', 'athlete_gender']) ['athlete_average_speed' ].mean ()
```

```python
#what age groups are the best in the 50m Race (20 + races min)
```

```python
df2.query('race_length == "50mi"') \
    .groupby('athlete_age')['athlete_average_speed'] \
    .agg(['mean', 'count']) \
    .query('count >= 20') \
    .sort_values('mean', ascending=False)
```

```python
#what age groups are the worst in the 50m Race (20 + races min) (show 20)
```

```python
df2.query('race_length == "50mi"') \
    .groupby('athlete_age')['athlete_average_speed'] \
    .agg(['mean', 'count']) \
    .query('count >= 9') \
    .sort_values('mean', ascending=True)
```

```python
#Seasons for the data -› Slower in summer than winter?
#spring 3-5
#Summer 6-8
#fall 9-11
#winter 12 -2
#split between two decimals
```

```python
df2[ 'race_month'] = df2['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df2[ 'race_season'] = df2['race_month']. apply(lambda x: 'Winter'
if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```
